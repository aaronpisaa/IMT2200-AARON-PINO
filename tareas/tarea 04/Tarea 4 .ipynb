```python
"""BUSCAMOS LAS FILAS DUPLICADAS EN EXPEDICIONES"""

#   creamos una columna con el tag true o false, siendo true si la fila expid está duplicado
df_exped['Column_duplicada'] = df_exped.duplicated(subset=['expid'], keep=False)

#  filtramos  la columna unicamente por los valores q son verdaderos
df_duplicados = df_exped[df_exped['Column_duplicada'] == True]

#buscamos en la columna expid cuales son los valores duplicados (nos ayudamos del true)
columna_expid_duplicados = df_duplicados['expid']

#obtenemos la lista de valores únicos de esos 'expid' repetidos
expid_duplicadas = columna_expid_duplicados.unique()

# Opcional: Eliminar la columna auxiliar
df_exped = df_exped.drop(columns=['Column_duplicada'])
print(f"\n Resumen de las filas expid repetidas \n  \
      {expid_duplicadas}")
print("Estas celdas que eran parte del dataframe de expedicion se \n" 
"repiten porque hay un error en el dataset original, y por tal motivo \n" \
"los identifico")
```

```python
# Creamos claves compuestas para asegurar unicidad

#  1. Crear la clave compuesta en df_exped
df_exped['expid_nueva'] = df_exped['expid'] + '_' + df_exped['year'].astype(str)

# 2. Crear la clave compuesta en df_members
df_members['expid_nueva'] = df_members['expid'] + '_' + df_members['myear'].astype(str)
```
