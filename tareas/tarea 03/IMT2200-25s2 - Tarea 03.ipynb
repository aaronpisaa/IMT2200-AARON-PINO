```python
#* b. 쮺칩mo se distribuye el IVM seg칰n nivel de ense침anza y ruralidad?

import pandas as pd

# Diccionario de DataFrames por nivel
niveles = {
    "Parvularia": df_parvularia,
    "1춿 B치sico": df_1_basico,
    "2춿 B치sico": df_2_basico,
    "Media": df_media
}

# Lista para guardar resultados
resumen_ivm = []

# Calcular el promedio de IVM por ruralidad en cada nivel
for nombre, df in niveles.items():
    # Asegurarnos de usar la columna correcta (ajusta si tu columna se llama distinto)
    if 'IVM' in df.columns:
        col_ivm = 'IVM'
    elif 'IVM Establecimiento' in df.columns:
        col_ivm = 'IVM Establecimiento'
    else:
        # Buscar columna que empiece con "IVM"
        col_ivm = [c for c in df.columns if 'IVM' in c][0]

    # Agrupar por ruralidad
    tabla = df.groupby("DS_RURALIDAD")[col_ivm].agg(['count', 'mean', 'std']).reset_index()
    tabla["Nivel"] = nombre
    resumen_ivm.append(tabla)

# Unir todos los resultados en una sola tabla
resumen_ivm = pd.concat(resumen_ivm, ignore_index=True)

# Reordenar columnas
resumen_ivm = resumen_ivm[["Nivel", "DS_RURALIDAD", "count", "mean", "std"]]
resumen_ivm.rename(columns={
    "DS_RURALIDAD": "Ruralidad",
    "count": "N춿 Establecimientos",
    "mean": "IVM Promedio",
    "std": "Desviaci칩n Est치ndar"
}, inplace=True)

print(resumen_ivm)
```

```python
# respuestas
import pandas as pd

niveles = {
    "Parvularia": df_parvularia,
    "1춿 B치sico": df_1_basico,
    "2춿 B치sico": df_2_basico,
    "Media": df_media
}

# Lista para guardar resultados
resumen_dep = []

for nombre, df in niveles.items():
    # Detectar columna de IVM
    col_ivm = [c for c in df.columns if 'IVM' in c][0]
    
    # Agrupar por tipo de dependencia
    tabla = (
        df.groupby("DS_TIPO_DEPENDENCIA")[col_ivm]
          .agg(['count', 'mean', 'std'])
          .reset_index()
    )
    tabla["Nivel"] = nombre
    resumen_dep.append(tabla)

# Unir todo
resumen_dep = pd.concat(resumen_dep, ignore_index=True)

# Ordenar columnas y renombrar
resumen_dep = resumen_dep[["Nivel", "DS_TIPO_DEPENDENCIA", "count", "mean", "std"]]
resumen_dep.rename(columns={
    "DS_TIPO_DEPENDENCIA": "Tipo de Dependencia",
    "count": "N춿 Establecimientos",
    "mean": "IVM Promedio",
    "std": "Desviaci칩n Est치ndar"
}, inplace=True)

print(resumen_dep)
```

### 5.1 C치lculo de IVM promedio por comuna (0.8 pts)

Primero, para cada uno sus 4 DataFrames, calcule el promedio de la columna `IVM Establecimiento` seg칰n comuna. Almacene este valor en una nueva columna llamada `IVM Promedio {NIVEL DE ENSE칌ANZA}` (por ejemplo, para ense침anza parvularia, cree la columna `IVM Promedio Parvularia`).

Junte los 4 resultados que obtuvo en un solo DataFrame `ivm_comunal` que contenga:
* Comuna (nombre y ID)
* Regi칩n (ID)
* IVM Promedio Parvularia de la comuna
* IVM Promedio B치sica Primer Ciclo de la comuna
* IVM Promedio B치sica Segundo Ciclo de la comuna
* IVM Promedio Media de la comuna

Ponga siempre atenci칩n al tama침o de sus DataFrames: como estamos agrupando comunalmente, estos siempre debiesen tener una cantidad de filas igual a las comunas de Chile.

```python
import pandas as pd

# --- 1. Calcular promedios comunales para cada nivel ---

# Parvularia
ivm_parv = (
    df_parvularia
    .groupby(['ID_REGION_ESTABLE', 'ID_COMUNA_ESTABLE', 'DS_COMUNA_ESTABLE'], as_index=False)
    ['IVM Establecimiento']
    .mean()
    .rename(columns={'IVM Establecimiento': 'IVM Promedio Parvularia'})
)

# 1춿 B치sico (Primer ciclo)
ivm_1basico = (
    df_1_basico
    .groupby(['ID_REGION_ESTABLE', 'ID_COMUNA_ESTABLE', 'DS_COMUNA_ESTABLE'], as_index=False)
    ['IVM Establecimiento']
    .mean()
    .rename(columns={'IVM Establecimiento': 'IVM Promedio B치sica Primer Ciclo'})
)

# 2춿 B치sico (Segundo ciclo)
ivm_2basico = (
    df_2_basico
    .groupby(['ID_REGION_ESTABLE', 'ID_COMUNA_ESTABLE', 'DS_COMUNA_ESTABLE'], as_index=False)
    ['IVM Establecimiento']
    .mean()
    .rename(columns={'IVM Establecimiento': 'IVM Promedio B치sica Segundo Ciclo'})
)

# Media
ivm_media = (
    df_media
    .groupby(['ID_REGION_ESTABLE', 'ID_COMUNA_ESTABLE', 'DS_COMUNA_ESTABLE'], as_index=False)
    ['IVM Establecimiento']
    .mean()
    .rename(columns={'IVM Establecimiento': 'IVM Promedio Media'})
)

# --- 2. Unir los cuatro resultados comunales ---

# Usamos merges sucesivos, conservando todas las comunas (outer join)
ivm_comunal = (
    ivm_parv
    .merge(ivm_1basico, on=['ID_REGION_ESTABLE', 'ID_COMUNA_ESTABLE', 'DS_COMUNA_ESTABLE'], how='outer')
    .merge(ivm_2basico, on=['ID_REGION_ESTABLE', 'ID_COMUNA_ESTABLE', 'DS_COMUNA_ESTABLE'], how='outer')
    .merge(ivm_media, on=['ID_REGION_ESTABLE', 'ID_COMUNA_ESTABLE', 'DS_COMUNA_ESTABLE'], how='outer')
)

# --- 3. Revisar el tama침o y estructura ---
print(ivm_comunal.shape)
ivm_comunal.head()
```

### 5.2 Establecimientos por comuna y ruralidad (0.8 pts)

Ahora contaremos la cantidad de establecimientos por comuna seg칰n ruralidad. Como buscamos la cantidad total de establecimientos, consideramos establecimientos de todos los niveles educacionales vistos. Para esto, combine los 4 DataFrames que carg칩 y genere un nuevo DataFrame `estab_comunal` que contenga:

* Comuna (ID)
* Cantidad de establecimientos rurales
* Cantidad de establecimientos urbanos
* Cantidad total de establecimientos
* Porcentaje de establecimientos rurales (Establecimientos rurales / Total establecimientos * 100)

Tenga en cuenta de que pueden aparecer establecimientos duplicados al combinar estos 4 DataFrames (por ejemplo, establecimientos que imparten m칰ltiples niveles de ense침anza, desde 1췈 b치sico hasta 4췈 medio). Preoc칰pese de manejar estos casos antes de agrupar sus datos seg칰n ruralidad.

```python
# respuestas
import pandas as pd

# --- 1. Combinar los cuatro DataFrames ---
df_todos = pd.concat([df_parvularia, df_1_basico, df_2_basico, df_media], ignore_index=True)

# --- 2. Eliminar duplicados por establecimiento ---
# Usamos ID_RBD, ya que identifica un칤vocamente a cada establecimiento en el pa칤s
df_todos = df_todos.drop_duplicates(subset='ID_RBD')

# --- 3. Agrupar por comuna y ruralidad ---
# Contamos la cantidad de establecimientos por tipo de ruralidad (Urbano / Rural)
conteo = (
    df_todos
    .groupby(['ID_COMUNA_ESTABLE', 'DS_RURALIDAD'])
    ['ID_RBD']
    .nunique()   # n칰mero de establecimientos 칰nicos
    .unstack(fill_value=0)  # convierte Ruralidad en columnas
    .rename(columns={'Rural': 'Estab_Rurales', 'Urbano': 'Estab_Urbanos'})
    .reset_index()
)

# --- 4. Calcular totales y porcentajes ---
conteo['Total_Estab'] = conteo['Estab_Rurales'] + conteo['Estab_Urbanos']
conteo['Pct_Rurales'] = (conteo['Estab_Rurales'] / conteo['Total_Estab']) * 100

# --- 5. (Opcional) a침adir nombres de comuna y regi칩n ---
# Si los necesitas en el resultado final:
df_aux = df_todos[['ID_REGION_ESTABLE', 'ID_COMUNA_ESTABLE', 'DS_COMUNA_ESTABLE']].drop_duplicates()
estab_comunal = conteo.merge(df_aux, on='ID_COMUNA_ESTABLE', how='left')

# --- 6. Reordenar columnas para mayor claridad ---
estab_comunal = estab_comunal[
    ['ID_REGION_ESTABLE', 'ID_COMUNA_ESTABLE', 'DS_COMUNA_ESTABLE',
     'Estab_Rurales', 'Estab_Urbanos', 'Total_Estab', 'Pct_Rurales']
]

# --- 7. Revisar resultados ---
print(estab_comunal.shape)
estab_comunal.head(40)
```

Finalmente, vamos a generar el DataFrame `agg_comunal` con la informaci칩n de los dos previos DataFrames creados.
Combine la informaci칩n de las dos secciones anteriores para crear su base de datos unificada. Por 칰ltimo, agregue la columna `geometry` con la informaci칩n geogr치fica de cada comuna.

```python

# ----- si a칰n no lo hiciste, ya tienes estos dos DF de pasos anteriores -----
# ivm_comunal: promedios por comuna y nivel
# estab_comunal: conteos de establecimientos por comuna (rural/urbano, total y %)

# Asegura tipos y crea una llave estandarizada de comuna (string 5 d칤gitos, p.ej. '15101')
for df in [ivm_comunal, estab_comunal]:
    df['ID_COMUNA_ESTABLE'] = pd.to_numeric(df['ID_COMUNA_ESTABLE'], errors='coerce').astype('Int64')
    df['COMUNA_ID'] = df['ID_COMUNA_ESTABLE'].astype('Int64').astype('string').str.zfill(5)

# Unimos ambos resultados comunales
agg_comunal = (
    ivm_comunal
    .merge(
        estab_comunal[
            ['ID_COMUNA_ESTABLE','COMUNA_ID',
             'Estab_Rurales','Estab_Urbanos','Total_Estab','Pct_Rurales']
        ],
        on=['ID_COMUNA_ESTABLE','COMUNA_ID'],
        how='left'
    )
)

# Si ivm_comunal no tra칤a estas columnas, ord칠nalas ahora:
base_cols = ['ID_REGION_ESTABLE','ID_COMUNA_ESTABLE','COMUNA_ID','DS_COMUNA_ESTABLE']
niveles_cols = [
    'IVM Promedio Parvularia',
    'IVM Promedio B치sica Primer Ciclo',
    'IVM Promedio B치sica Segundo Ciclo',
    'IVM Promedio Media'
]
rural_cols = ['Estab_Rurales','Estab_Urbanos','Total_Estab','Pct_Rurales']

cols_finales = base_cols + niveles_cols + rural_cols
agg_comunal = agg_comunal[cols_finales]
```

```python
import pandas as pd
import geopandas as gpd

# --- Leer shapefile ---
gdf = gpd.read_file("zip://comunas.zip").to_crs(epsg=4326)

# --- Crear COMUNA_ID ---
gdf['COMUNA_ID'] = (
    pd.to_numeric(gdf['cod_comuna'], errors='coerce')
      .astype('Int64')
      .astype('string')
      .str.zfill(5)
)

# --- Prepara geometry ---
gdf_geo = gdf[['COMUNA_ID', 'geometry']].drop_duplicates('COMUNA_ID')

# --- Crear COMUNA_ID tambi칠n en tus DataFrames tabulares ---
for df in [ivm_comunal, estab_comunal]:
    df['COMUNA_ID'] = (
        pd.to_numeric(df['ID_COMUNA_ESTABLE'], errors='coerce')
          .astype('Int64')
          .astype('string')
          .str.zfill(5)
    )

# --- Unir todo ---
agg_comunal = (
    ivm_comunal
    .merge(
        estab_comunal[
            ['COMUNA_ID','Estab_Rurales','Estab_Urbanos','Total_Estab','Pct_Rurales']
        ],
        on='COMUNA_ID',
        how='left'
    )
    .merge(gdf_geo, on='COMUNA_ID', how='left')
)

# --- Convertir a GeoDataFrame final ---
agg_comunal = gpd.GeoDataFrame(agg_comunal, geometry='geometry', crs='EPSG:4326')

# --- Revisar ---
print("九 Filas:", len(agg_comunal))
print("九 Columnas:", agg_comunal.columns.tolist())
print("九 Comunas sin geometr칤a:", agg_comunal['geometry'].isna().sum())
agg_comunal.head(20)

```
