#### (*6 pts*) 1.4.6 Considerando su nuevo DataFrame ¿hay alguna relación entre la distancia viajada y el número de pasajeros? 

Para contestar esta pregunta realice un gráfico de distancia viajada vs. cantidad de pasajeros (gráfico de disperción). No olvide poner título al gráfico y a los ejes, y comentar lo que infiere usted al ver el gráfico.

Respuesta:

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)

ax.plot(viajes_selec['trip_distance'], viajes_selec['passenger_count'],'.')
ax.set_xlabel('Distancia')
ax.set_ylabel('Pasajeros')
ax.set_title('Distancia de viaje vs pasajeros')
```

En este gráfico de dispersión no se observa alguna relación clara que haya entre el número de pasajeros y la distancia recorrida. Es importante notar que no tenemos certeza de si realmente hay o no una relación, pues para ello debemos hacer un análisis estadístico de los datos (test de hipótesis) y por otro lado el gráfico de dispersión con números discretos de pasajeros no nos permite ver la acumulación de viajes en ciertas distancias. Para ver mejor la distribución deberíamos hacer otro tipo de gráfico como veremos más adelante en el ramo en la parte de visualización. 

Por ejemplo, podríamos hacer un gráfico de violín para identificar las distribuciones como se ve en el siguiente ejemplo:

```python
pasajeros = []
for p in range(1, viajes_selec['passenger_count'].max()+1):
    trips_por_pasajero = viajes_selec[(viajes_selec['passenger_count']==p)]
    pasajeros.append(trips_por_pasajero['trip_distance'])
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)

ax.violinplot(pasajeros, showmeans = True,)
ax.set_xlabel('Pasajeros')
ax.set_ylabel('Distancia')
ax.set_title('Distribución de distancia según número de pasajeros')
```
