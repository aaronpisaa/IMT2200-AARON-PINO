#### (*6 pts*) 1.4.6 Considerando su nuevo DataFrame hay alguna relacin entre la distancia viajada y el nmero de pasajeros? 

Para contestar esta pregunta realice un grfico de distancia viajada vs. cantidad de pasajeros (grfico de dispercin). No olvide poner ttulo al grfico y a los ejes, y comentar lo que infiere usted al ver el grfico.

Respuesta:

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)

ax.plot(viajes_selec['trip_distance'], viajes_selec['passenger_count'],'.')
ax.set_xlabel('Distancia')
ax.set_ylabel('Pasajeros')
ax.set_title('Distancia de viaje vs pasajeros')
```

En este grfico de dispersin no se observa alguna relacin clara que haya entre el nmero de pasajeros y la distancia recorrida. Es importante notar que no tenemos certeza de si realmente hay o no una relacin, pues para ello debemos hacer un anlisis estadstico de los datos (test de hiptesis) y por otro lado el grfico de dispersin con nmeros discretos de pasajeros no nos permite ver la acumulacin de viajes en ciertas distancias. Para ver mejor la distribucin deberamos hacer otro tipo de grfico como veremos ms adelante en el ramo en la parte de visualizacin. 

Por ejemplo, podramos hacer un grfico de violn para identificar las distribuciones como se ve en el siguiente ejemplo:

```python
pasajeros = []
for p in range(1, viajes_selec['passenger_count'].max()+1):
    trips_por_pasajero = viajes_selec[(viajes_selec['passenger_count']==p)]
    pasajeros.append(trips_por_pasajero['trip_distance'])
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)

ax.violinplot(pasajeros, showmeans = True,)
ax.set_xlabel('Pasajeros')
ax.set_ylabel('Distancia')
ax.set_title('Distribucin de distancia segn nmero de pasajeros')
```
